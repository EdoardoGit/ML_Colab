```python
CATEGORIE = ("Dogs/","Cats/")
DIR='/content/ML_Colab/CatsDogs/'
IMG_SIZE = 75
```

```python
from IPython.display import Image
Image('/content/ML_Colab/CatsDogs/Cats/7585.jpg')
```

```python
def create_training_data():
  training_data = []
  for categories in CATEGORIE:
    path = os.path.join(DIR,categories)
    print(path)
    class_num = CATEGORIE.index(categories)
    for image in tqdm(os.listdir(path)):
      try:
        img_array = cv2.imread(os.path.join(path,image),cv2.IMREAD_GRAYSCALE)
        #print(img_array.shape)
        #if img_array is None:
        #  print('Wrong path:', os.path.join(path,image))
        #else:
        new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))

        training_data.append([new_array,class_num])
      except Exception as e:
        print(e)
        pass
  return training_data
```

```python
data=create_training_data()
print(len(data))
```

```python
random.shuffle(data)
print(len(data))
```

```python
def data_construct(data):
  input=[]
  output=[]

  for img,label in data:
    input.append(img)
    y=np.zeros(len(CATEGORIE))
    y[label]=1
    output.append(y)
  input=np.array(input).reshape(-1,IMG_SIZE,IMG_SIZE,1)
  return input,output
```

```python
x,y = data_construct(data)
```

```python
i = 0
x_p1 = []
x_p2 = []
for data in x:
  if(i<len(x)/2):
    x_p1.append(data)
  else:
    x_p2.append(data)
  i=i+1
```

```python
import pickle

pickle_out = open("Gray_Shuffled_Data_p1.pickle","wb")
pickle.dump(x_p1,pickle_out)
pickle_out.close()

pickle_out = open("Gray_Shuffled_Data_p2.pickle","wb")
pickle.dump(x_p2,pickle_out)
pickle_out.close()

pickle_out = open("Gray_Shuffled_Labels.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```
